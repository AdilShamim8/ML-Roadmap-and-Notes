```python
def run (batsman_name):
    vk = df[df['batsman'] == batsman_name]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).index[0]
```

```python
run('CH Gayle')
```

***isin()***  
***Find the most destructive death over batsman in the history of IPL***  
***Strike Rate = (Number of runs / Number of balls) * 100***  
***Min batsman 200 balls in over 16-20***

```python
# Identify death overs (overs greater than 15)  
death_over = df[df['over'] > 15]  
```

```python
# Group by batsman and count the number of runs scored  
all_batsman = death_over.groupby('batsman')['batsman_runs'].count()
x =all_batsman > 200
batsman_list =all_batsman[x].index.tolist()
```

```python
# Runs scored by all these 43 batsman  
# Balls played by these 43 batsman  

final = df[df['batsman'].isin(batsman_list)]  
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum()  
```

```python
balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
sr = (runs/balls)*100
```

#### Applying Pivot Table to IPL Dataset

```python
match = pd.read_csv('matches.csv')
delivery = pd.read_csv('deliveries.csv')
```

```python
# Create a mask for deliveries with 6 runs  
mask = delivery['batsman_runs'] == 6  

# Filter the DataFrame to include only deliveries with 6 runs  
six = delivery[mask]  

# Display the shape of the filtered DataFrame  
print(six.shape)  
```

```python
# Create a pivot table with 'over' as index and 'batting_team' as columns  
pivot_table = six.pivot_table(index='over', columns='batting_team', values='batsman_runs', aggfunc='count') 
pivot_table
```

```python
import seaborn as sns
```

```python
sns.heatmap(pivot_table)
```
